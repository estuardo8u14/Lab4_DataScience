```python
f = open('en_US.blogs.txt', 'r',encoding="utf8")
blogs = f.readlines()
f.close()
```

```python
f = open('en_US.twitter.txt', 'r',encoding="utf8")
tweets = f.readlines()
f.close()
```

```python
f = open('en_US.news.txt', 'r',encoding="utf8")
news = f.readlines()
f.close()
```

```python
###SE eliminaron las mayusculas
```

```python
tweets= [i.lower() for i in tweets]
news= [i.lower() for i in news]
blogs= [i.lower() for i in blogs]
tweets
```

```python
###Se eliminaran los URLS
```

```python
tweets=[re.sub('\w+:\/{2}[\d\w-]+(\.[\d\w-]+)*(?:(?:\/[^\s/]*))*','',i) for i in tweets]
news=[re.sub('\w+:\/{2}[\d\w-]+(\.[\d\w-]+)*(?:(?:\/[^\s/]*))*','',i) for i in news]
blogs=[re.sub('\w+:\/{2}[\d\w-]+(\.[\d\w-]+)*(?:(?:\/[^\s/]*))*','',i) for i in blogs]
```

```python
###eliminar caracteres especiales
```

```python
tweets=[re.sub('[^a-zA-Z0-9 ]+','', i) for i in tweets]
news=[re.sub('[^a-zA-Z0-9 ]+', '', i) for i in news]
blogs=[re.sub('[^a-zA-Z0-9 ]+', '', i) for i in blogs]
```

```python
##Eliminar Stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
stopwords=set(stopwords.words('english'))
trendy_words=['im','rt','jk','btw','lol','yolo','lmao','lmfao','fb','like','get']
for i in trendy_words:
    stopwords.add(i)
```

```python
##mostrar archivos limpios
```

```python
tweets_clean=[]
for tweet in tweets:
    word_list=[]
    for word in tweet.split():
        if word not in stopwords:
            word_list.append(word)
    tweets_clean.append(' '.join(word_list))
tweets_clean
```

```python
blogs_clean=[]
for blog in blogs:
    word_listo=[]
    for word in blog.split():
        if word not in stopwords:
            word_listo.append(word)
    blogs_clean.append(' '.join(word_listo))
blogs_clean
```
